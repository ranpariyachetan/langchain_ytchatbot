```python
from youtube_transcript_api import YouTubeTranscriptApi

ytt_api = YouTubeTranscriptApi()
```

```python
video_id = 'hF-7eKtzAHM'  # Replace with your video ID

transcripts = ytt_api.get_transcript(video_id)

# Print the transcript
print(transcripts[100])
```

```python
transcripts_text = ' '.join([t["text"] for t in transcripts])
print(transcripts_text)
```

```python
transcripts_text = str.replace(transcripts_text, '\n', ' ')
transcripts_text
```

```python
transcripts_text = str.replace(transcripts_text, '  ', ' ')
transcripts_text = str.replace(transcripts_text, '♪', '')
transcripts_text
```

```python
print(len(transcripts_text))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len,
)

transcript_chunks = text_splitter.create_documents([transcripts_text])

print(transcript_chunks[0])
print(len(transcript_chunks))
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from dotenv import load_dotenv
load_dotenv()
embedding = OpenAIEmbeddings()

chroma_store = Chroma.from_documents(
    transcript_chunks,
    embedding,
    persist_directory="chroma_store",
    collection_name="youtube_transcript",
)
```

```python
chroma_store.similarity_search("What is the main topic of the video?", k=3)
```

```python
retriever = chroma_store.as_retriever(search_type="similarity", search_kwargs={"k": 4})

retriever
```

```python
retriever.invoke("What is James Webb Telescope?")
```

```python
from langchain.prompts import PromptTemplate

prompt_template = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    You are helpful assistant.
    Answer ONLY from the provided transcript context.
    If the context is not sufficient to answer the question, say "I don't know".

    {context}
    Question: {question}
    """,
)
```

```python
question = "What is James Webb Telescope?"
retrieved_docs = retriever.invoke(question)
print(retrieved_docs[0].page_content)
```

```python
context_text = "\n\n".join([doc.page_content for doc in retrieved_docs])
final_prompt = prompt_template.invoke({"context": context_text, "question": question})

final_prompt
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI()

answer =  llm.invoke(final_prompt)

print(answer)
```
